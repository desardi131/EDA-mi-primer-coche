```python
from bs4 import BeautifulSoup as bs
import requests
import pandas as pd
import numpy as np
```

```python
url = 'https://neomotor.epe.es/conduccion/las-marcas-y-modelos-de-coche-mas-fiables-del-mercado-EJNM8886'
response = requests.get(url)
```

```python
html = response.content
```

```python
soup = bs(html, "lxml")
```

```python
soup
```

```python
title = soup.find_all('ol')[1].find_all('li')[0].get_text().split(':')[0]#[1:3]
title
```

```python
puntos = soup.find_all('ol')[1].find_all('li')[0].get_text().split(':')[1][1:3]
puntos
```

```python
data = {'title' : [], 'puntos' : []}

for i in range(36):
    data['title'].append(soup.find_all('ol')[1].find_all('li')[i].get_text().split(':')[0])
    data['puntos'].append(soup.find_all('ol')[1].find_all('li')[i].get_text().split(':')[1][1:3])

data
```

```python
df_fiabilidad = pd.DataFrame(data)
```

```python
df_fiabilidad
```

```python
df_fiabilidad.to_csv('fiabilidad.csv', index = False)
```
